```python
import pycocotools
import cv2
```

```python
from pycocotools.coco import COCO
import os
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
coco = COCO('/content/valid/annotation/_annotations.coco.json')
img_dir = '/content/valid'

for image_id in range(0,482):
  img = coco.imgs[image_id]
  cat_ids = coco.getCatIds()
  anns_ids = coco.getAnnIds(imgIds=img['id'], catIds=cat_ids, iscrowd=None)
  if (anns_ids):
    anns = coco.loadAnns(anns_ids)
    mask = coco.annToMask(anns[0])
    for i in range(len(anns)):
        mask += coco.annToMask(anns[i])
    plt.imshow(mask)
    path = '/content/Mask_valid'
    path2 = '/content/Rail_valid'
    pth = '/content/valid'
    Rail = cv2.imread(os.path.join(pth,img['file_name']))

    cv2.imwrite(os.path.join(path2,'Rail{}.png'.format(image_id)),Rail)
    cv2.imwrite(os.path.join(path,'mask{}.png'.format(image_id)),mask*255)
```
